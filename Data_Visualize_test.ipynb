```python
#Scatter Plot
import h5py
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

h5path='ply_data_test2.h5'
pt_f = h5py.File(h5path, 'r')
data = pt_f['data'][:]



ax.scatter(data[0, 0:100, 0],data[0, 0:100, 1], data[0, 0:100, 2],c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
